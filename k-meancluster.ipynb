1. It is unsupervised learning algortihm.
2. It is used to identity groups and find target audience.

Characterstics of groups:-
1. Each group has a distinct quality
2. No group overlaps each other.
3. THe total number of groups are decided by using the Elbow method
4. Min number of groups = 2
5. Max num of groups = Num of Observation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('./files/income.csv')
```

```python

df.head(22)
```

```python
df.info()
```

```python
df.drop(columns = "Name", inplace = True)
```

```python
plt.scatter(df["Age"], df["Income($)"])
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=3, random_state= 55)
km.fit(df)
```

```python
km.predict(df)
```

```python
df["cluster"] = km.predict(df)
```

```python
df
```

```python
df1 = df[df["cluster"] == 0]
df2 = df[df["cluster"] == 1]
df3 = df[df["cluster"] == 2]
```

```python
df1
```

```python
plt.scatter(df1["Age"], df1["Income($)"])
plt.scatter(df2["Age"], df2["Income($)"])
plt.scatter(df3["Age"], df3["Income($)"])
```

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler()
```

```python
df["Age"] = sc.fit_transform(df[["Age"]])
df["Income($)"] = sc.fit_transform(df[["Income($)"]])
```

```python
km1 = KMeans(n_clusters=3, random_state=5)
km1.fit(df)
df["cluster"] = km1.predict(df)
```

```python
df1 = df[df["cluster"] == 0]
df2 = df[df["cluster"] == 1]
df3 = df[df["cluster"] == 2]
```

```python
plt.scatter(df1["Age"], df1["Income($)"])
plt.scatter(df2["Age"], df2["Income($)"])
plt.scatter(df3["Age"], df3["Income($)"])
```
